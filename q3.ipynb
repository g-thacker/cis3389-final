```python
import pandas as pd
```

```python
df = pd.read_csv('Titanic.csv')
```

```python
df.shape
```

```python
grouped = df['Fare'].groupby(df['Pclass'])
grouped.agg(['mean','std'])
```

```python
df = df.dropna()
df.shape
```

```python
# 183 rows remain
```

```python
df['Survived'] = df['Survived'].map({0: 'No', 1: 'Yes'})
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(df['Gender'], df['Age'], color = 'yellow')
plt.title('Age by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
```

```python
# The male age bar is higher.
```

```python
plt.bar(df['Survived'], df['Age'], color = 'green')
plt.title('Age by Survival')
plt.xlabel('Survived')
plt.ylabel('Age')
```

```python
# Older passengers had a higher likelihood of survival.
```
